##**Imporing Pandas Library**

```python
import pandas as pd
```

##**Loading Datasets in pandas**

```python
df = pd.read_csv('/content/sample_data/california_housing_test.csv')
df
```

```python
print(df.columns)
```

##**Groupby & Aggregate**

```python
df.groupby('longitude').mean()
```

```python
df.groupby('median_house_value').count().head(10)
```

```python
minimum_median_house_value = df.groupby('median_house_value').min()
minimum_median_house_value
```

```python
maximum_median_house_value = df.groupby('median_house_value').max()
maximum_median_house_value
```

```python
sum_median_house_value = df.groupby('median_house_value').sum()
sum_median_house_value
```

```python
df.groupby('median_house_value').describe()
```

##**Aggregating in Pandas**

```python
df.groupby('households').agg(['count', 'min', 'max' , 'sum' , 'mean']).head(10)
```

```python
df.groupby('households').agg({'households' : ['count', 'min', 'max' , 'sum' , 'mean']}).head(3)
```

```python
df.groupby('households').describe()
```
